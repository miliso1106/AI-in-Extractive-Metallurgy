```python
import json
with open('/content/ml_models_by_process.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
print("JSON loaded successfully")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

rows = []
for process_name, info in data["processes"].items():
    rows.append({
        "processName": process_name,
        "r2": info.get("metrics", {}).get("r2"),
        "mae": info.get("metrics", {}).get("mae"),
        "best_candidate": info.get("best_candidate"),
    })

dfp = pd.DataFrame(rows)

# R2 graph
plt.figure(figsize=(12,6))
sns.barplot(data=dfp.sort_values("r2", ascending=False), x="r2", y="processName", color="steelblue")
plt.title("R2 by Process")
plt.tight_layout()
plt.show()

# MAE graph
plt.figure(figsize=(12,6))
sns.barplot(data=dfp.sort_values("mae", ascending=True), x="mae", y="processName", color="tomato")
plt.title("MAE by Process")
plt.tight_layout()
plt.show()
```
